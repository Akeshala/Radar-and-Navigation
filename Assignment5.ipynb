Assumptions:

$d \gg R$ 

$|A(t)|\:= 1$ for signals $A_1$ and $A_2$

for $1 \gg |x|$ we can assume that $\sqrt{1-x} \approx 1-\frac{x}{2}$ and $\sqrt{1+x} \approx 1+\frac{x}{2}$

From cosine rule, $d_1$ and $d_2$ can be calculated as follows. 

\begin{align}
d_1 &= \sqrt{R^2+d^2-2dR\cos(2\pi f_mt-\theta)}\\
&= d\sqrt{\frac{R^2}{d^2}+1-2\frac{R}{d}\cos(2\pi f_mt-\theta)}\\
&= d\sqrt{1-2\frac{R}{d}\cos(2\pi f_mt-\theta)} \\
d_1&=d-R\cos(2\pi f_mt-\theta)\\
Assumptions, d \gg R \implies \frac{R^2}{d^2} \approx 0\\ \sqrt{1-x} \approx 1-\frac{x}{2} \\
d_2 \text{ can be obtained similarly,}
\\
d_2&=d+R\cos(2\pi f_mt-\theta)
\end{align}

\begin{align}
V_{A1}&=\sin(2\pi (f_c+f_0)t+\phi_1)\\

\end{align}

\begin{align}

By taking $V_{A0}$ as our reference, signal relative phase shift can be obtained as follows. 
\begin{align}

\Delta\phi_2&= \frac{2\pi}{\lambda}(d2-d)&=-\frac{2\pi}{\lambda}R\cos(2\pi f_mt-\theta) &=-\Delta\phi\\
\end{align}

We can simplify this as,
\begin{align}
V_{R_x}&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct \right)+\sin(2\pi (f_c+f_0)t+\Delta\phi) \\&+\sin(2\pi (f_c-f_0)t-\Delta\phi)\\
&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct \right)+2\sin(2\pi f_ct)\cos(2\pi f_mt +\Delta\phi )\\
&=(1-M\sin(2\pi f_mt)+2\cos(2\pi f_mt +\Delta\phi ))\sin\left(2\pi f_ct \right)\\
V_{R_x}&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_mt +\frac{2\pi R}{\lambda}\cos(2\pi f_mt-\theta) \right)\right\}\sin\left(2\pi f_ct \right)

```python
import math
import seaborn as sns; 
from scipy.signal import freqz,hilbert
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_c = 1000000 # Carrier frequency 
f_0 = 960 # Frequency offset
f_m = 30 # Rotational frequency
d = 200000 # Distance to the target
theta = (( 2*math.pi )/360)*30 # Phase angle from magnetic North 
M = 0.3 # Modulation index
delta = 16 
C = 300000000
```

```python
l = C/f_c # wave length
R = 8*l/math.pi # radius of antenna 

w_0 = 2*math.pi*f_0
w_m = 2*math.pi*f_m
w_c = 2*math.pi*f_c

F_s = int(10*f_c) # sampling frequency
E = 1.0/M
freq_div = delta * f_m
phase_shift = (2*math.pi*d)/l

# creating time axis
t = np.linspace(0, 1, F_s + 1)

print('Phase shift due to distance -', str(phase_shift), 'rad')
print('Radius of antenna circle -', str(R), 'm')
```

```python
d_1 = d - R*np.cos(w_m*t-theta)
d_2 = d + R*np.cos(w_m*t-theta)

delt_phi_1 = (2*math.pi/l)*d_1
delt_phi_2 = (2*math.pi/l)*d_2

V_A0 = E*(1-M*np.sin(w_m*t))*np.sin(w_c*t-phase_shift)
V_A1 = E*np.sin((w_c+w_0)*t - delt_phi_1)
V_A2 = E*np.sin((w_c-w_0)*t - delt_phi_2)

plt.figure(figsize=(20,5))
plt.plot(t[:300000], V_A0[:300000])
plt.title('Recieved signal from antenna A0')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.show()

plt.figure(figsize=(20,5))
plt.plot(t[0:3000], V_A1[0:3000])
plt.ylim((-4 ,4))
plt.title('Recieved signal from antenna A1')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.show()

plt.figure(figsize=(20,5))
plt.plot(t[0:3000], V_A2[0:3000])
plt.ylim((-4 ,4))
plt.title('Recieved signal from antenna A2')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.show()
```

## Support functions

```python
def plot_FFT(abs_FFT,F_s,range,name):
  N = len(abs_FFT)
  freq_axis = np.fft.fftfreq(N, 1/F_s)
  plt.figure(figsize=(25, 5))
  ax=plt.ylabel('$|V(f)|^2$')
  ax=plt.xlabel('Frequency')
  ax=plt.title('Frequency response of -'+name)
  plt.xlim(range)
  ax=plt.plot(freq_axis,  (2/N)*abs_FFT)
  return None

def LPF(cutoff,fs,N,plot,name):
  H = np.zeros((1,N))
  H[0,0:cutoff+1] = np.ones((1,cutoff+1))
  H[0,-cutoff:] = np.ones((1,cutoff))
  H=H[0]
  if plot:
    plt.figure(figsize=(25,5))
    ax=plt.ylabel('$|H(f)|^2$')
    ax=plt.xlabel('Frequency')
    ax=plt.title('Lowpass filter for -'+name)
    plt.plot(H,color='r')
    plt.xlim((0,cutoff+50))
  return H

def BPF(cutoff_L,cutoff_R,fs,N,plot,name):
  L_idx = cutoff_L
  R_idx = cutoff_R
  H = np.zeros((1,N))
  H[0,cutoff_L:cutoff_R+1 ]= np.ones((1,cutoff_R-cutoff_L+1))
  H[0,-cutoff_R-1:-cutoff_L]=np.ones((1,cutoff_R-cutoff_L+1))
  H = H[0]
  if plot:
    plt.figure(figsize=(25,5))
    ax=plt.ylabel('$|H(f)|^2$')
    ax=plt.xlabel('Frequency')
    ax=plt.title('Banpass filter for -'+name)
    plt.plot(H,color='r')
    plt.xlim((0,cutoff_R+50))
  return H   
```

# Question 5

Rotation discussed in above steps can be electronically simulated by a circular array of antenna elements. Cosine blending must be used here while transmitting. 

Fixed number of omni directional antennas are arranged in the circle with the radius $R$ of previous setup. Rotation can be simulated by activating and deactivating radial pairs of the antenna setup.

Since radiation cannot be switched abruptly, zero switching concept has been introduced. Cosine blending function is used to ramping up and down the power each transmitter.

Adding antenna pairs into odd and even groups and handing over will be done from odd to odd and even to even.

\begin{align}
f&=\frac{2\pi f_m}{4\times(2\pi / \text{antenna_count})}\\
&=\frac{2\pi \times 30}{4\times(2\pi /48)}\\
&=360Hz
\end{align}

Blending odd and even function generation

\begin{align}
\text{cosine blending function for odd antennas}&= |\sin(2\pi ft)|\\
\text{cosine blending function for even antennas}&= |\cos(2\pi ft)|
\end{align}

```python
antennas=48
blend_freq= (np.pi*f_m)/(2*(2*np.pi / antennas))
pulse_width= 2/(f_m*antennas)
print("Blending frequency-", blend_freq)
```

```python
halfw_sampled=int(0.5*pulse_width*F_s)+1
sample_period=int(F_s/f_m)+1
rep=31
beta=360/antennas

t_arr=np.arange(0,2/f_m,1/F_s)
```

```python
discr_sin=np.sin(blend_freq*t_arr)
discr_sin[2*halfw_sampled-1:sample_period]=0
odd_func=np.array(list(discr_sin[:sample_period])*rep)

discr_cos=np.cos(blend_freq*t_arr)
discr_cos[halfw_sampled-1:sample_period-halfw_sampled]=0
even_func=np.array(list(discr_cos[:sample_period])*rep)
```

Generate the composite received signal with fixed antenna array. Antenna rotation simulation can be done as follows. 

```python
V_Rx=0
for i in range(antennas):
  if i==0:
    angle=beta/2
    pulse=even_func
  else:
    angle+=beta
    pulse=np.array(list(np.zeros(((i-1)*halfw_sampled)))+list(odd_func))
  blend=pulse[0:N]
  dx=R*np.cos((theta-angle)*np.pi/180)
  phi_1=2*np.pi*(d-dx)/l
  phi_2=2*np.pi*(d+dx)/l

  upper_sp=E*np.sin(2*np.pi*(f_c+f_0)*t-phi_1)*blend
  lower_sp=E*np.sin(2*np.pi*(f_c-f_0)*t-phi_2)*blend
  V_Rx+=upper_sp+lower_sp
```
